```python
import pandas as pd
```

```python
#df = pd.read_csv("Year_wise_batting_stats_all_ODI_players(20-12-19).csv")
df = pd.read_csv("best-batting-average.csv")
df = df.fillna(0)
df
```

```python
#data = df[df["country"]=="IND"]
#data
```

```python
data = df[df["Year"]>=2014]
data
```

```python
fdata = data[data["Inns"]>=15]
fdata
```

```python
fdata = fdata.fillna(0)
fdata  
```

```python
x= fdata.iloc[:,[3,6,8,10]].values
y =fdata.iloc[:,[2]].values
```

```python
from sklearn.preprocessing import StandardScaler

```

```python
a =6
sc=StandardScaler()
x=sc.fit_transform(x)

#predicting the test set resuts
#y_pred=classfier.predic

from sklearn.svm import SVC
df=SVC(kernel='rbf',random_state=0)
df.fit(x,y)
y_pred=df.predict(x)

s= []
for i in y_pred:
    if i not in s:
        s.append(i)
batsman=[]

for i in range(0,a):
    batsman.append(s[i])
f = pd.DataFrame(batsman, columns = ['Batsman Name']) 
f
```

```python
bb = pd.read_csv("best-bowling-economy-innings (1).csv")
```

```python
bb = bb[bb["Year"]>=2014]
bb = bb.dropna()
bb
```

```python

d= bb.iloc[:,[7,5,4,3]].values
e= bb.iloc[:,[2]].values
e
```

```python
from sklearn.preprocessing import StandardScaler
c=6
sc=StandardScaler()
d=sc.fit_transform(d)
 
#predicting the test set resuts
#y_pred=classfier.predic

from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
classifier.fit(d,e)
e_pred=classifier.predict(d)                       
                         
e_pred
```

```python
f = []
for i in e_pred:
    if i not in f:
        f.append(i)
bowler=[]
for i in range(0,c): 
    bowler.append(f[i])    
g = pd.DataFrame(bowler, columns = ['Bowler Name']) 
g
```
