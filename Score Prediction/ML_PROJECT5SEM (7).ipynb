```python
import pandas as pd
wc = pd.read_csv('ODI_match.csv')
#wc = wc[wc['bat_team']=="United Arab Emirates"]
#wc = wc[wc['bowl_team']=="West Indies"]
dataset = pd.read_csv('odi.csv')
dataset = dataset[dataset['bat_team']=="India"]
dataset = dataset[dataset['bowl_team']=="England"]
dataset = dataset[dataset['date']=="2013-01-11"]
wc
```

```python
X = wc.iloc[:,[7,8,9,12,13]].values #Input features
X
```

```python
y = wc.iloc[:, 14].values #Label
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
ODIsc = StandardScaler()
X_train = ODIsc.fit_transform(X_train)
X_test = ODIsc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
score = lr.score(X_test,y_test)*100
print("R-squared value:" , score)
```

```python
def accuracy(y_test,y_pred,frequency):
    r = 0
    l = len(y_pred)
    for i in range(0,l):
        if(abs(y_pred[i]-y_test[i]) <= frequency):
            r= r+1
    return ((r/l)*100)
```
